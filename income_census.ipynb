```
import pandas as pd
import censusdata
import psycopg2
from sqlalchemy import create_engine

engine = create_engine('postgresql://naya:urigoren@104.155.40.113:5432/nrich')
```

```
income = censusdata.download('acs5', 2015, censusdata.censusgeo([('place','*')]),
                    ['B06010_001E','B06010_002E','B06010_004E','B06010_005E','B06010_006E','B06010_007E','B06010_008E','B06010_009E','B06010_010E','B06010_011E'])

income.columns=['Total number of people', 'No income','$1 to $9,999 ','	$10,000 to $14,999','	$15,000 to $24,999','$25,000 to $34,999','$35,000 to $49,999','	$50,000 to $64,999','$65,000 to $74,999','incom75K_or_more']
income.describe()
income
```

```
income.to_sql("im_income_by_place", engine ,if_exists='replace',index=False)
```

```
# value_list= [str(k).split(',')[2].split(',')[0][6:] for k in income.index.values]
income_dict={}
for k in income.index.values:
  a = str(k).split(',')
  address=a[0]
  state=a[1].split(':')[0].strip()
  place=a[2].split('place:')[-1]
  
  if state not in income_dict:
     income_dict[state]=[]
    
  income_dict[state].append([address, place])
  
income_dict
```

```
income.to_sql("im_income_by_place", engine ,if_exists='replace',index=False)
```
